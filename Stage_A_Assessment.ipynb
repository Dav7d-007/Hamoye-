```python
#Loading the Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv',encoding='latin-1')
df.describe(include = 'all')
```

```python
#Question 2
#Total Sum of Wine produced in 2015 and 2018 respectively?

df[df['Item'] == 'Wine'].groupby('Item')[['Y2015','Y2018']].sum()
```

```python
#Question 3
#What is the total number of unique countries in the dataset?
df['Area'].nunique()
```

```python
#Question 4
#Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.
#Which of these Areas had the 7th lowest sum in 2017?
df_group = df[['Y2017', 'Area']].groupby('Area')
df_sum = df_group['Y2017'].sum()
seventh_lowest_area = df_sum.sort_values(ascending=True).index[6]
print("Area with the 7th lowest sum in 2017:", seventh_lowest_area)
```

```python
#Question 6
#What is the total number and percentage of missing data in 2014 to 3 decimal places?
total = df['Y2014'].count()
total_2014 = df['Y2014'].isnull().sum()
print("The total number and percentage of missing data in 2014 is:", total_2014, round((total_2014/total)*100, 3),'%')
```

```python
#Question 14
#Perform a groupby operation on ‘Element’.  What is the total number of the sum of Processing in 2017?
total_processing_2017 = df[df['Element'] == 'Processing'].groupby('Element')['Y2017'].sum()
print("Total Processing in 2017:", total_processing_2017.iloc[0])
```

```python
#Question 15
#Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.
#Which of these Areas had the highest sum in 2017?

df_highest = df.groupby('Area')[['Area','Y2017']].sum().sort_values(by='Y2017',  ascending=False).head()
area_with_highest_sum = df_highest.max()
print("Area with the highest sum in 2017:", area_with_highest_sum)
```

```python
#Question 16
#What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?
mean_2017 = df['Y2017'].mean().round(2)
std_2017 = df['Y2017'].std().round(2)
print("Mean for 2017:", mean_2017, "\nStandard Deviation for 2017:", std_2017)
```
